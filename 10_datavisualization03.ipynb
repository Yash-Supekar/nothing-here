```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv(url, header=None, names=columns)

# Inspect the first few rows of the dataset
print(iris.head())
```

```python
# Get the data types of each feature
print(iris.dtypes)

# List of features and their types
features_types = {
    "sepal_length": "numeric",
    "sepal_width": "numeric",
    "petal_length": "numeric",
    "petal_width": "numeric",
    "species": "nominal"
}
print(features_types)
```

```python
# Create histograms for each feature
iris_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for feature in iris_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(iris[feature], kde=True, bins=20, color='blue')
    plt.title(f"Histogram of {feature}")
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Create boxplots for each feature
for feature in iris_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='species', y=feature, data=iris, palette="Set2")
    plt.title(f"Boxplot of {feature} by Species")
    plt.xlabel('Species')
    plt.ylabel(feature)
    plt.show()
```
